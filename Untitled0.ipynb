**SCANIA Component X Dataset**

The dataset has 9 CSV files total, split across 3 sets (Train / Validation / test)



*   Training Set (3 files)
    1) train_operational_readouts.csv - This is main sensor data. Each row is one readout from one vehicle at a specific time. Contain vehicle_id, time_step, and 105 sensor columns (6 histogram variables with multiple bins each, 8 numerical counters).  

    2) train_tte.csv - Time to event file. One row per vehicle. Contains vehicle_id, length_of_study_time_step (how long the vehicle was observed), and in_study_repair (0 = no failure, 1 = Component X was repaired/failed). This is your training label file.

    3) train_specifications.csv - One row per vehicle. Contains vehicle_id plus 8 categorical specification columns (Spec_0 to Spec_7) Describing things like engine type and wheel configuration. All anonymized (values are Cat0, Cat1, … Cat28).

*   Validation Set (3 Files)
    1) validation_operational_readouts.csv - Same column structure ad the training readouts, but each vehicles history is cut off at a randomly selected readout.

    2) validation_labels.csv - One row per vehicle. Contains vehicle_id and class_label (0-4), where the class tells you which time window before failure the last readout fell in: 0 = healthy (>48 units away), 1 = 24–48 units, 2 = 12–24 units, 3 = 6–12 units, 4 = 0–6 units (imminent).

    3) validation_specifications.csv - Same structure as train specification, for the 5046 validation vehicles.

*   Test Set (3 files)
    1) test_operational_readouts.csv - Same as validation readouts but for the 5045 test vehicles.

    2) test_labels.csv - Same structure as validation labels, for the 5045 test vehicles.

    3) test_specifications.csv - Same structure as train/validation specifications, for the test vehicles.

**Class distribution — validation & test**
1. Printing counts and percentages
   *  Loops through validation and test sets
   *  Counts sample in each class
   *  Prints both absolute counts and percentage of split

2. Visualization
   *  Creates two subplots: one for validation, one for test
   *  Use bar chart with color-coded class
   *  Applies logarithmic y-axis to handle class imbalance
   *  Annotates bars for clarity
   *  Saves figure to PNG
  
3. Colors & labels
   *  Consistent color scheme for classes
   *  Short x-axis labels to fit nicely in plots
   


```python
print("\n" + "=" * 60)
print("VALIDATION & TEST — CLASS LABEL (0–4)")
print("=" * 60)

class_meaning = {               #Define meaning for each class label (0 = Healthy, 4 = imminent failure)
    0: "> 48 time units before failure (healthy)",
    1: "24–48 time units before failure",
    2: "12–24 time units before failure",
    3: "6–12 time units before failure",
    4: "0–6 time units before failure (imminent)",
}

for split_name, labels_df in [("Validation", val_labels), ("Test", test_labels)]:
    print(f"\n  {split_name}:")                  #Loop into validation and test dataset to print counts and Percentages per class

    vc2 = labels_df["class_label"].value_counts().sort_index()  #count samples in each class and sort by class label (0 -> 4)

for cls, cnt in vc2.items():
        pct = cnt / len(labels_df) * 100  # Calculate percentage of total samples in split
        # Print formatted summary: class, meaning, count, percentage
        print(f"    Class {cls} — {class_meaning[cls]:<42}  n={cnt:>4,}  ({pct:.2f}%)")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))    #Create figure with two subplots side by side
colors = ["#4C72B0", "#55A868", "#C44E52", "#8172B2", "#DD8452"]    #defines colors for the 5 classes for consistent visual distinction
class_labels = ["0\n>48", "1\n24-48", "2\n12-24", "3\n6-12", "4\n0-6"]     #define x-axis labels for bars

for ax, (split_name, labels_df) in zip(axes, [("Validation", val_labels), ("Test", test_labels)]):  #loops through subplots and dataset splits
    vc2 = labels_df["class_label"].value_counts().sort_index()  #count sample per class and sort by class label

    #create bar chart for the split
    bars = ax.bar(class_labels, #x axis label
                  vc2.values,  #y axis values
                  color=colors, #bar colors
                  edgecolor="white", #white border around bars
                  linewidth=0.8) #thickness of edge

    #set titles and labels for axes
    ax.set_title(f"{split_name} set — class distribution")
    ax.set_xlabel("Class (time window before failure)")
    ax.set_ylabel("Number of vehicles")

    #use log scale on y axis to better viaualize imbalances classes
    ax.set_yscale("log")

    #format y axis as standard numbers
    ax.yaxis.set_major_formatter(mticker.ScalarFormatter())

    #annotate each bar with the exact count of vehicles
    for bar, val in zip(bars, vc2.values):
        ax.text(bar.get_x() + bar.get_width() / 2,     #x position: center of the bar
                bar.get_height() * 1.1,                #y position: sightly above the bar
                str(val),                              #text: count value
                ha="center", va="bottom", fontsize=9)  #horizontal alignment and vertical alignment

plt.tight_layout()  #adjust layout to prevent overlapping elements
plt.savefig("fig_class_distribution_val_test.png", bbox_inches="tight") #save figure as PNG
plt.show()  #display the figure
```

**Cost matrix Heatmap**

Cost Matrix:  Misclassification cost matrix

Heatmap: Visualizes this matrix so you can quickly see which errors are most expensive


```python
import numpy as np
cost_matrix = np.array([
    [0,   7,   8,   9,  10],
    [200, 0,   7,   8,   9],
    [300, 200, 0,   7,   8],
    [400, 300, 200, 0,   7],
    [500, 400, 300, 200, 0],
])

#diagnoal = correct predictions (cost = 0)
#off diagonal = penalties for misclassification

fig, ax = plt.subplots(figsize=(7, 5)) #figure and axis for the heatmap

mask = np.eye(5, dtype=bool) #mask to hide the diagonal

sns.heatmap(
    cost_matrix,          # Data to visualize
    annot=True,
    fmt="d",              # Format as integer
    cmap="YlOrRd",        # Color map: yellow → red
    mask=mask,
    linewidths=0.5,
    ax=ax,
    xticklabels=[f"Pred {i}" for i in range(5)],
    yticklabels=[f"Actual {i}" for i in range(5)]
)

ax.set_title("Misclassification cost matrix\n(diagonal = correct prediction, cost = 0)")
ax.set_xlabel("Predicted class")
ax.set_ylabel("Actual class")

plt.tight_layout()
plt.savefig("fig_cost_matrix.png", bbox_inches="tight")
plt.show()
print("Key insight: missing a class-4 failure (predicting 0) costs 500 — 50× more than a false alarm.")
```

**Operational Data of Column Structure**

Analyze operational data columns and categorize them into:

Histogram Variables -> Features split into multiple bins (sensor reading over time or ranges)

Counter variables -> single column feature (not split)

```python
print("\n" + "=" * 60)
print("OPERATIONAL DATA — COLUMN STRUCTURE")
print("=" * 60)

feature_cols = [c for c in train_op.columns if c not in ["vehicle_id", "time_step"]] #extract feature columns
print(f"  Total feature columns : {len(feature_cols)}")

# Identify histograms vs counters by checking multi-bin variables
from collections import defaultdict

var_bins = defaultdict(list)

for col in feature_cols:       #loops through all feature columns
    parts = col.rsplit("_", 1)
    if len(parts) == 2 and parts[1].isdigit():
        var_bins[parts[0]].append(int(parts[1]))
    else:
        var_bins[col].append(0)

histograms = {k: sorted(v) for k, v in var_bins.items() if len(v) > 1}
counters   = {k: v for k, v in var_bins.items() if len(v) == 1}

print(f"\n  Histogram variables   : {len(histograms)}")
for var, bins in sorted(histograms.items()):
    print(f"    {var:<12}  {len(bins)} bins  →  columns: {var}_{bins[0]} … {var}_{bins[-1]}")

print(f"\n  Counter variables     : {len(counters)}")
for var in sorted(counters.keys()):
    print(f"    {var}")
```

**Missing Value analysis**

```python
print("\n" + "=" * 60)
print("MISSING VALUES — OPERATIONAL DATA")
print("=" * 60)

def missing_summary(df, name):
    miss = df[feature_cols].isnull().mean() * 100
    print(f"\n  {name}:")
    print(f"    Max missing in any column  : {miss.max():.3f}%")
    print(f"    Mean missing across columns: {miss.mean():.4f}%")
    print(f"    Columns with any missing   : {(miss > 0).sum()}")
    return miss

miss_train = missing_summary(train_op, "Train operational")
miss_val   = missing_summary(val_op,   "Validation operational")
miss_test  = missing_summary(test_op,  "Test operational")

fig, ax = plt.subplots(figsize=(14, 4))
ax.bar(range(len(miss_train)), miss_train.values, color="#4C72B0", alpha=0.7, label="Train")
ax.bar(range(len(miss_val)),   miss_val.values,   color="#DD8452", alpha=0.5, label="Validation")
ax.set_title("Missing value percentage per feature column (train set)")
ax.set_ylabel("Missing (%)")
ax.set_xlabel("Feature column index")
ax.set_ylim(0, 1.5)
ax.axhline(1.0, color="red", linestyle="--", linewidth=0.8, label="1% threshold")
ax.legend()
plt.tight_layout()
plt.savefig("fig_missing_values.png", bbox_inches="tight")
plt.show()


```
